```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
# creating an empty list
xlst = []

# number of elements as input
x = int(input("Enter number of coordinates : "))

# iterating till the range
for i in range(0, x):
    x = int(input())
    # adding the element
    xlst.append(x)
x = np.array(xlst)
x
```

```python
# creating an empty list
ylst = []

# number of elements as input
y = int(input("Enter number of coordinates (Enter same number as x) : "))

# iterating till the range
for i in range(0, y):
    y = int(input())
    # adding the element
    ylst.append(y)
y = np.array(ylst)
y
```

```python
# creating an empty list
zlst = []

# number of elements as input
z = int(input("Enter number of coordinates (Enter same number as x): "))

# iterating till the range
for i in range(0, z):
    z = int(input())
    # adding the element
    zlst.append(z)
z = np.array(zlst)
z
```

```python
cells = 50
nCPTS = np.size(x, 0)
n = nCPTS - 1
i = 0
u = np.linspace(0, 1, cells)
b = []
```

```python
xBezier = np.zeros((1, cells))
yBezier = np.zeros((1, cells))
zBezier = np.zeros((1, cells))
```

```python
def comb(n, i):
  return np.math.factorial(n) / (np.math.factorial(i) * np.math.factorial(n-i))
```

```python
def bernstein(n, i, u):
  J = np.array(comb(n, i) * (u**i) * ((1-u) ** (n-i)))
  return J
```

```python
for k in range(0, nCPTS):
  b.append(bernstein(n, i, u))

  xBezier = bernstein(n, i, u) * x[k] + xBezier
  yBezier = bernstein(n, i, u) * y[k] + yBezier
  zBezier = bernstein(n, i, u) * z[k] + zBezier
  i+=1
```

```python
for line in b:
  plt.plot(u, line)
plt.show()
```

```python
fig = plt.figure(figsize=(5, 5))
ax1 = fig.add_subplot(111, projection='3d')
ax1.scatter(x, y, z, c='black')
ax1.plot(xBezier[0], yBezier[0], zBezier[0], c='red')
plt.show()
```
